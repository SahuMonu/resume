```python
!pip install telethon
!pip install pytz
import os
import re
import requests
import asyncio
from telethon import TelegramClient, events
from telethon.errors import FloodWaitError
import nest_asyncio
from urllib.parse import urlparse, parse_qs, urlencode, urlunparse
from datetime import datetime
import pytz

# Apply the nest_asyncio patch to allow running event loops in Colab
nest_asyncio.apply()

# Replace with your actual API ID, API Hash, and phone number
api_id = '27143743'
api_hash = '7a90866e1527c4243354c32d9b722e50'
phone_number = '+917974089097'
session_name = 'my_telegram_user_session'

# Replace with your actual Amazon affiliate tag
affiliate_tag = 'techstore0380-21'  # Your Amazon affiliate tag

# Replace with your target channel link or username where you want to send the message
target_channel = "https://t.me/hugediscoutandoffer"

# List of source channels to listen to
source_channels = [
     "https://t.me/botpractice123456789", # Test Channel
    "https://t.me/+sX1Ht4p33nFjZDE1", # Offerzone 2.0
    "https://t.me/Deal_Bazaar_Coupons_Discounts",#Deals Bazaarâš¡ï¸
    "https://t.me/+tcpHvG6zIWs2MmJl", #Hot Deals
    "https://t.me/+eHrweX4qZg5jNzI9", #Nctricks
    "https://t.me/+KgUrCwnDny02ZDk1",#Deals point
    "https://t.me/+lEEQraQyKwljNGQ1", #Offer Box 2.0
    "https://t.me/+7rHGiXC8TnsxN2I9", #Offerzone
    "https://t.me/+TejaYfLClkAJ5WD4", # IND Loot Deals
    "https://t.me/+Th6aG5Zaxz_i_u7a"#Offer Box Official
]

# Keywords and links to skip
skip_keywords = ["join", "telegram", "join our group", "join our channel"]
telegram_link_pattern = re.compile(r'https?://t\.me/[\w]+', re.IGNORECASE)

# Step 1: Delete the previous session file if it exists
if os.path.exists(f"{session_name}.session"):
    os.remove(f"{session_name}.session")
    print(f"Deleted previous session file: {session_name}.session")

# Step 2: Initialize the Telegram client using the API ID and API Hash
client = TelegramClient(session_name, api_id, api_hash)

def expand_shortened_url(shortened_url):
    """Expand a shortened URL to its full URL."""
    try:
        response = requests.head(shortened_url, allow_redirects=True)
        return response.url
    except requests.RequestException as e:
        print(f"Error expanding URL {shortened_url}: {e}")
        return shortened_url

def clean_url(url):
    """Remove 'linkID', 'linkCode' and 'tag' parameters from the URL and print them."""
    parsed_url = urlparse(url)
    query_params = parse_qs(parsed_url.query)

    # Extract and print 'linkID', 'linkCode' and 'tag' parameters
    link_id = query_params.get('linkId', [None])[0]
    link_code = query_params.get('linkCode', [None])[0]
    tag = query_params.get('tag', [None])[0]

    if link_id:
        print(f"Found linkID: {link_id}")
    if link_code:
        print(f"Found linkCode: {link_code}")
    if tag:
        print(f"Found tag: {tag}")

    # Remove 'linkID', 'linkCode' and 'tag' from query parameters
    query_params.pop('linkId', None)
    query_params.pop('linkCode', None)
    query_params.pop('tag', None)

    # Add the new affiliate tag
    query_params['tag'] = affiliate_tag

    # Rebuild the query string without 'linkID' and 'linkCode'
    new_query_string = urlencode(query_params, doseq=True)

    # Reconstruct the URL without the removed parameters
    cleaned_url = urlunparse(parsed_url._replace(query=new_query_string))

    return cleaned_url

async def handle_flood_wait_error(func):
    """Handle FloodWaitError and retry after waiting."""
    while True:
        try:
            await func()
            break
        except FloodWaitError as e:
            wait_time = e.seconds
            print(f"Flood wait error. Sleeping for {wait_time} seconds.")
            await asyncio.sleep(wait_time)

# Define the IST timezone
IST = pytz.timezone('Asia/Kolkata')

# Store the start time in IST when the program starts
start_time = datetime.now(IST)

# Function to calculate the time difference and return a message with the timestamp
def add_timestamp(message: str) -> str:
    """
    This function appends the current IST timestamp to a message and calculates
    the total time difference in hours and minutes since the program started.

    Parameters:
    - message (str): The message to which the timestamp is to be added.

    Returns:
    - str: The message with the IST timestamp and total time difference appended.
    """
    # Get current time in IST
    now = datetime.now(IST)
    # Calculate the time difference between now and the start time
    time_difference = now - start_time

    # Extract the total hours and minutes from the time difference
    hours, remainder = divmod(time_difference.total_seconds(), 3600)
    minutes, _ = divmod(remainder, 60)


    # Format the timestamp as 'YYYY-MM-DD HH:MM:SS' in IST
    timestamp = now.strftime("%Y-%m-%d %H:%M:%S")
    message = "The bot is running."
    # Return the message with the IST timestamp and time difference
    return f"{timestamp} IST - {message} \n (Time since start: {int(hours)} : {int(minutes)} )"


async def main():
    # Start the client and authenticate with the phone number
    await client.connect()

    # Check if already authorized
    if not await client.is_user_authorized():
        # Send code request
        await handle_flood_wait_error(lambda: client.send_code_request(phone_number))

        # Sign in the user
        await client.sign_in(phone_number, input('Enter the code you received: '))

    # Get the target channel entity
    target_channel_entity = await client.get_entity(target_channel)

    # List to store cleaned links
    cleaned_links_list = []

    # Function to handle new messages
    @client.on(events.NewMessage(chats=source_channels))
    async def handle_new_message(event):
        message = event.message
        message_text = message.text

        # Skip messages containing certain keywords or Telegram links
        if any(keyword in message_text.lower() for keyword in skip_keywords) or telegram_link_pattern.search(message_text):
            print(f"Skipping message containing unwanted content from {event.chat.title}:")
            print(f"Sender ID: {message.sender_id}")
            return  # Skip processing this message

        # Define a regex pattern to match both standard and shortened Amazon URLs
        amazon_link_pattern = re.compile(
            r'https?://(?:www\.)?(?:amzn\.to|amazon\.(?:com|co\.uk|de|fr|it|es|ca|in|jp|au|mx|br|pl|tr|nl|se|no|dk|fi|cz|hu|ro|sg|ae|sa|kw|eg|ar|cl|co|pe|ve|uy|za))/[^\s]*',
            re.IGNORECASE
        )

        # Find all Amazon links in the message
        amazon_links = amazon_link_pattern.findall(message_text)
        if amazon_links:
            print(f"New message containing Amazon link(s) from {event.chat.title}:")
            print(f"Sender ID: {message.sender_id}")
            modified_message_text = message_text
            for link in amazon_links:
                # Expand shortened URLs
                expanded_link = expand_shortened_url(link)
                # Clean URL by removing 'linkID', 'linkCode' and 'tag'
                cleaned_link = clean_url(expanded_link)
                print(f"Original Link: {link}")
                print(f"Expanded Link: {expanded_link}")
                print(f"Cleaned Link: {cleaned_link}")
                print(add_timestamp(message))

                # Check for duplicates
                if cleaned_link in cleaned_links_list:
                    print(f"Duplicate link found, skipping both the original and cleaned links.")
                    return  # Exit the function to skip sending this message

                # Add the cleaned link to the list
                cleaned_links_list.append(cleaned_link)

                # Replace the original link in the message with the cleaned link
                modified_message_text = modified_message_text.replace(link, cleaned_link)
                print('---')

            # Send the message to the target channel if no duplicates were found
            await handle_flood_wait_error(lambda: client.send_message(target_channel_entity, modified_message_text))
            print(f"Sent modified message to target channel: {modified_message_text}")
            print(f"Cleaned Links List: {cleaned_links_list}")

    print(f"Listening for new messages in {', '.join(source_channels)}...")

    # Run the client until it is manually stopped
    await client.run_until_disconnected()

# Run the main function using asyncio's event loop
asyncio.run(main())
```
